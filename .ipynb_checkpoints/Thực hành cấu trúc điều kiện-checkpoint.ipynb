### Bài 3: Viết chương trình nhập vào 3 số a, b, c và in ra số lớn nhất


````python
a = int(input("Input number a: "))
b = int(input("Input number b: "))
c = int(input("Input number c: "))

if a > b and a > c:
    print(f"Max is {a}")
elif b > c:
    print(f"Max is {b}")
else: 
    print(f"Max is {c}")
````

### Bài 4: Viết chương trình nhập vào số tuổi của 3 người và cho biết ai là người trẻ nhất và ai là người lớn tuổi nhất

````python
num1 = int(input("Nhập số tuổi của người thứ nhất: "))
num2 = int(input("Nhập số tuổi của người thứ hai: "))
num3 = int(input("Nhập số tuổi của người thứ ba: "))
max_age = num1
max_index = 1
if max_age < num2:
    max_age = num2
    max_index = 2
if max_age < num3:
    max_age = num3
    max_index = 3
    
min_age = num1
min_index = 1
if min_age > num2:
    min_age = num2
    min_index = 2
if min_age > num3:
    min_age = num3
    min_index = 3
    
print(f"Người thứ {max_index} là người lớn tuổi nhất ({max_age} tuổi)")
print(f"Người thứ {min_index} là người nhỏ tuổi nhất ({min_age} tuổi)")
````

### Bài 6: Học viên sẽ không đủ điều kiện thi nếu tỉ lệ có mặt nhỏ hơn 75%. Viết chương trình nhập vào tổng số buổi học của một khóa học và tổng số buổi học mà học viên tham gia. Sau đó cho biết học viên đó có đủ điều kiện thi hay không.


````python
total_lessons = int(input("Input total lessons: "))
attended_lessons = int(input("Input attended lessons: "))

if attended_lessons / total_lessons >= 0.75:
    print("You can join the final exam")
else:
    print("You cannot join the final exam")
````

### Bài 7: Một công ty vận tải có cách tính chi phí giao hàng như sau:
1. 10kilogram đầu tiên: \$5/kg
2. Từ kilogram thứ 11 đến 20: \$7/kg
3. Từ kilogram thứ 20 trở lên: \$10/kg

Viết chương trình nhập vào số kg cần giao(số nguyên) và in ra chi phí giao hàng tương ứng


````python
goods_weight = int(input("Nhập khối lượng món hàng: "))
delivery_cost = 0

if goods_weight <= 10:
    delivery_cost = 5 * goods_weight
elif 10 < goods_weight <= 20:
    delivery_cost = 5 * 10 + 7 * (goods_weight - 10)
else: 
    delivery_cost = 5 * 10 + 7 * 10 + 10 * (goods_weight - 20)
    
print(f"Chi phí giao hàng là {delivery_cost}")
````

### Bài 8: Viết chương trình nhập vào số năm và cho biết năm đó là năm nhuận hay năm không nhuận.

````python
year = int(input("Nhập số năm: "))
if not year % 400:
    print(f"Năm {year} là năm nhuận")
elif not year % 4 and year % 100:
    print(f"Năm {year} là năm nhuận")
else:
    print(f"Năm {year} không phải là năm nhuận")
````

### Bài 10:
Tại một cửa hàng nọ có menu như sau:

1. Gà rán giá 35k
2. Hamburger giá 50k
3. Pesi giá 10k

Nhập số lượng món hàng cần mua

Tính và in ra tổng giá tiền

Nếu giá trị đơn hàng trên 2 triệu, giảm 10% cho khách, trên 4 triệu, giảm 15%

Tính và in ra tổng giá tiền được giảm

Tính và in ra tổng giá tiền khách cần trả

In ra dòng chữ "Cảm ơn quý khách đã mua hàng"


````python
"""Ví dụ:
Nhập số lượng gà rán khách mua:100
Nhập số lượng hamburger khách mua:100
Nhập số lượng pesi khách mua:15
Tổng giá tiền đơn hàng là: 8.650.000
Tổng tiền được giảm là: 1.297.500
Tổng tiền phải thanh toán: 7.352.500
Cảm ơn quý khách đã mua
"""
````

````python
chicken = int(input("Nhập số lượng gà rán khách mua:"))
hamburger = int(input("Nhập số lượng hamburger khách mua:"))
pepsi = int(input("Nhập số lượng pepsi khách mua:"))
total_price = chicken * 35000 + hamburger * 50000 + pepsi * 10000
print(f"Tổng giá tiền đơn hàng là: {total_price}")
if total_price > 4000000:
    print(f"Tổng tiền được giảm là: {total_price*0.15}")
    print(f"Tổng tiền phải thanh toán: {total_price*0.85}")
elif total_price > 2000000:
    print(f"Tổng tiền được giảm là: {total_price*0.1}")
    print(f"Tổng tiền phải thanh toán: {total_price*0.9}")
else:
    print(f"Tổng tiền phải thanh toán: {total_price}")
````

### Bài 11:

Nhập vào chiều cao (cm) và cân nặng (kg), tính số BMI và xét rồi xin kết quả theo dữ liệu sau:

- BMI < 16: Gầy cấp độ III
- 16 <= BMI < 17:  Gầy cấp độ II
- 17<= BMI < 18.5: Gầy cấp độ I
- 18.5 <= BMI < 25: Bình thường
- 25 <= BMI < 30: Thừa cân
- 30 <= BMI < 35 : Béo phì cấp độ I
- 35 <= BMI < 40: Béo phì cấp độ II
- BMI > 40: Béo phì cấp độ III


````python
height = int(input("Nhập chiều cao(m): "))
weight = int(input("Nhập cân nặng(kg): "))
bmi = weight / height ** 2
if bmi < 16:
    print("Gầy cấp độ III")
elif bmi < 17:
    print("Gầy cấp độ II")
elif bmi < 18.5:
    print("Gầy cấp độ I")
elif bmi < 25:
    print("Bình thường")
elif bmi < 30:
    print("Thừa cân")
elif bmi < 35:
    print("Béo phì cấp độ I")
elif bmi < 40:
    print("Béo phì cấp độ II")
else:
    print("Béo phì cấp độ III")
````

### Bài 12: Tìm ra người lớn tuổi nhất trong 4 người thông qua số tuổi nhập vào từ người dùng.
